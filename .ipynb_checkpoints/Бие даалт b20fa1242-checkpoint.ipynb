```python
# 
del seminar ["No"]
seminar["Нийт -25"]= seminar.apply(lambda row: sum(row==1)/ 19*25 ,axis=1)
irts["Нийт -10"]= irts.apply(lambda row: sum(row=='и')/32*10 ,axis=1)
seminar["Нийт -25"]= seminar.apply(lambda row: sum(row==1)/ 19*25 ,axis=1)
shalgalt = shalgalt.assign(Нийт=lambda x: (x['Шалгалт - 1'] + x['Шалгалт - 2'] + x['Шалгалт - 3']))
bie_daalt= bie_daalt.assign(Нийт=lambda x: (x['Бие даалт - 1'] + x['Бие даалт - 2']))
humuujil_handlaga= humuujil_handlaga.assign(Нийт=lambda x: (x['ХХ-1'] + x['ХХ-2'] + x['ХХ-3'] + x['ХХ-4']))
nemelt_onoo= nemelt_onoo.assign(Нийт=lambda x: (x['Нэмэлт оноо']))
# 

niit['Ирц-10'] = irts.iloc[: , -1]
niit['Семинар-25'] = seminar.iloc[: , -1]
niit['Шалгалт-30'] = shalgalt.iloc[: , -1]
niit['Бие даалт-25'] = bie_daalt.iloc[: , -1]
niit['ХХ-10'] = humuujil_handlaga.iloc[: , -1]
niit['Нэмэлт оноо'] = nemelt_onoo.iloc[: , -1]
del niit['Unnamed: 10']
niit = niit.drop(28)

# niit['Нийт'] = niit[['Ирц-10','Семинар-25','Шалгалт-30','Бие даалт-25','ХХ-10','Нэмэлт оноо']].sum(axis=1)
# niit
niit['Нийт'] = niit.loc[:,'Ирц-10': 'Нэмэлт оноо'] .sum(axis=1)
niit.set_index('No')
niit
```

```python
def unelgee(x):
    if x>=96:
        return "A"
    elif x>=90 and x<=95:
        return "A-"
    elif x>=87 and x<=89:
        return "B+"
    elif x>=83 and x<=86:
        return "B"
    elif x>=80 and x<=82:
        return "B-"
    elif x>=77 and x<=79:
        return "C+"
    elif x>=73 and x<=76:
        return "C"
    elif x>=70 and x<=72:
        return "C-"
    elif x>=67 and x<=69:
        return "D+"
    elif x>=63 and x<=66:
        return "D"
    elif x>=60 and x<=62:
        return "D-"
    else:
        return "F"
niit['Үнэлгээ'] = round(niit['Нийт']).apply(unelgee)
niit
```

```python
niit['unelgee_too'] = 1
niit.groupby(['Үнэлгээ'])['unelgee_too'].count()
```

```python
niit_amjilt = niit.loc[niit['Үнэлгээ'] != 'F', 'unelgee_too'].sum()
amjilt = niit_amjilt / 28 * 100
print('Хичээлийн амжилт:', round(amjilt,1), '%')
```

```python
niit_chanar = niit.loc[niit['Үнэлгээ'].isin(['A', 'B','A-','B+','B-']), 'unelgee_too'].sum()
chanar = niit_chanar / 28 * 100
print('Хичээлийн чанар:', round(chanar,2), '%')
```

```python
F_dundaj = niit.loc[niit['Үнэлгээ'] == 'F','Нийт'].mean()
F_max = niit.loc[niit['Үнэлгээ'] == 'F', 'Нийт'].max()
F_min = niit.loc[niit['Үнэлгээ'] == 'F', 'Нийт'].min()

print('F дундаж дүн:', round(F_dundaj,2))
print('F хамгийн их:' , round(F_max,2))
print('F хамгийн бага:', round(F_min,2))
```
